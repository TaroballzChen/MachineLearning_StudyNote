```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
```

```python
wine.data
```

```python
wine.target
```

```python
wine.data.shape
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
Xtrain.shape
```

```python
tree_clf = tree.DecisionTreeClassifier(criterion="entropy")
tree_clf = tree_clf.fit(Xtrain,ytrain)
score = tree_clf.score(Xtest,ytest)
```

```python
score
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(tree_clf
                                ,feature_names=wine.feature_names
                               , class_names=["琴酒","雪莉","貝爾摩德"]
                               ,filled=True   # 是否上色
                               ,rounded=True  # 方塊是否圓滑
                               )
graph = graphviz.Source(dot_data)
```

```python
import matplotlib.pyplot as plt
y_eff = []
for i in range(10):
    tree_clf = tree.DecisionTreeClassifier(criterion="entropy"
                                          ,random_state = 30
                                          ,splitter = "random"
                                          ,max_depth = i+1
                                          )
    tree_clf = tree_clf.fit(Xtrain,ytrain)
    score = tree_clf.score(Xtest,ytest)
    y_eff.append(score)
```

```python
plt.plot(range(1,11),y_eff,color="red",label="max_depth")
plt.legend()
plt.show()
```

```python
y_eff = []
for i in range(20):
    tree_clf = tree.DecisionTreeClassifier(criterion="entropy"
                                            ,random_state = 30
                                            ,splitter = "random"
                                            ,max_depth = 5
                                            ,min_samples_leaf=i+1
                                          )
    tree_clf = tree_clf.fit(Xtrain,ytrain)
    score = tree_clf.score(Xtest,ytest)
    y_eff.append(score)
plt.plot(range(1,21),y_eff,color="red",label="min_samples_leaf")
plt.legend()
plt.show()
```
